### Merge all data points

```python
frames = [Healthy,OHDC,OHDNC,betaOHD,betaHSD,betaHSDlike,POR,Aldosteron,OHD,NonCAH]
Disease_Data = pd.concat(frames)
Disease_Data.count()
```

### Convert categorical string column to one hot encoded format

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = pd.get_dummies(Disease_Data['Gender'], drop_first=True)
# Drop column non_numeric_col as it is now encoded
Disease_Data = Disease_Data.drop('Gender',axis = 1)
# Join the encoded dataframe to X
# Disease_Data = Disease_Data.join(onehotencoder)
# Disease_Data[Male] = onehotencoder['M']
onehotencoder.head()

Disease_Data['Male'] = onehotencoder['M']
```

```python
## Create a function to convert non numerical values or bad datapoints to Null
import re

def remove_characters(x):
    regex = re.compile('[@_!#$%^&*()<>?/\|}{~:]') 
    if str(type(x))=="<class 'str'>":
        if(regex.search(x) == None): 
            return x  
        elif(regex.search(x) == '*'):
            return re.sub('[^A-Za-z0-9]+', '', x)
        else:            
            return re.sub('[^A-Za-z0-9]+', '', x )
    else:
        return x
    
```

```python
# As per part 1 of the assignment convert the target column into two categories - 'Healthy' and 'Has Disease'
# Combining all disease into 'Has Disease'

def health_indicator(x):
    if x == 'Healthy':
        return('Healthy')
    else:
        return('Has Disease')
```

```python
Disease_Data['Health_Indicator'] = Disease_Data.apply(lambda row: health_indicator(row.Disease) , axis=1)
```

```python
Disease_Data.head()
```

```python
Disease_Data.head()
```

```python
drop_columns = ['Disease', 'Health_Indicator', 'PatientID']
X= Disease_Data.drop(drop_columns, axis = 1)
y= Disease_Data['Health_Indicator']
strat = Disease_Data['Disease']
```

### Data Cleanup 

```python
X.applymap(remove_characters)
```

```python
# Replacing * with blank
X.replace('*','')
```

```python
X.replace('<', '')
X.replace('>', '')
X.replace(',', '')
```

```python
for column in X:
#     X[i] = pd.to_numeric(X[i], errors ='coerce').fillna(0).astype('float')
    X[column]=X[column].apply(lambda x: x if np.isreal(x) else 0)
```

```python
X.replace([np.inf, -np.inf, np.nan], 0, inplace=True)
```

# Build prediction models

### Splitting data into training and test set for running Machine Learning algorithm

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = strat, test_size = 0.30)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression(random_state = 0)
lr_clf.fit(X_train, y_train)
```

```python
print('Accuracy on training set:',lr_clf.score(X_train,y_train))
print('Accuracy on test set:',lr_clf.score(X_test,y_test))
```

```python
from sklearn.model_selection import cross_val_score
# clf = svm.SVC(kernel='linear', C=1)
scores_lr = cross_val_score(lr_clf, X_train, y_train, cv=5)
scores_lr
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores_lr.mean(), scores_lr.std() * 2))
```
